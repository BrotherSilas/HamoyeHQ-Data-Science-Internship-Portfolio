```python
#data wrangling
import pandas as pd
import csv
url='https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel_data=pd.read_csv(url)
fuel_data.head()
```

```python
fuel_data.describe(include='all')
```

```python
#checking for missing values
fuel_data.isnull().sum()
```

```python
#using groupby to count the sum of each unique value in the fuel unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
```

```python
fuel_data[['fuel_unit']]=fuel_data[['fuel_unit']].fillna(value='mcf')
#checking if missing values have been filled
fuel_data.isnull().sum()
#yes, they have been filled
```

```python
#grouping by fuel type code year 
fuel_data.groupby('report_year')['report_year'].count()
```

```python
#and print the first entries in all the groups formed
fuel_data.groupby('fuel_type_code_pudl').first()
```

```python
#splitting fuel_data into two groups for merging using different methods
fuel_data1=fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_data2=fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
#checking that the length of both dataframes sum to the expected length
assert len(fuel_data)==(len(fuel_data1) + len(fuel_data2))
```

```python
#inner merge will lose rows that dont match in both dataframes
#returns rows available or present in both dataframes
pd.merge(fuel_data1, fuel_data2, how="inner")
```

```python
#outer merge will return all rows in both dataframes
pd.merge(fuel_data1, fuel_data2, how="outer").head()
```

```python
#left merge will remove rows from the right dataframe
#that do not have a match in the left dataframe
pd.merge(fuel_data1, fuel_data2, how="left")
```

```python
#concatenation, to join fuel_data2 to fuel_data1
#pd.concat([fuel_data,data_to_concat]).reset_index(drop=True)
pd.concat([fuel_data1,fuel_data2]).reset_index(drop=True)
```
